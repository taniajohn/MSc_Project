```python
import json

# Load JSON objects from file
json_objects = []
with open('/kaggle/input/amazondataset/amazon_gng_food.json', 'r') as file:
    for line in file:
        json_objects.append(json.loads(line))
```

```python
import pandas as pd

# Convert the list of JSON objects to a DataFrame
df = pd.DataFrame(json_objects)
```

```python
df.info()
```

```python
# Extract only required columns
selected_columns = ["category", "description", "title","asin"]
df = df[selected_columns]
```

```python
df.head()
```

```python
# Randomly select 10000 rows from the dataset
amazon_dataset_10000 = df.sample(n=10000, random_state=42)
```

```python
amazon_dataset_10000.to_csv('amazon_dataset_10000.csv', index=False)
```

```python
# Extract unique categories under 'Grocery and Gourmet Foods'
unique_second_items = df['category'].apply(lambda categories_list: categories_list[1] if len(categories_list) >= 2 else None).unique().tolist()

print(unique_second_items)
```

```python
# Filter rows where the second item in the category column is 'Breakfast Foods'
filtered_rows = df[df['category'].apply(lambda categories_list: len(categories_list) >= 2 and categories_list[1] == 'Breakfast Foods')]
```

```python
# Randomly select 50 rows from the filtered DataFrame
random_selection = filtered_rows.sample(n=50, random_state=42)
```

```python
# Save the dataset as a CSV file
random_selection.to_csv('amazon_dataset_new.csv', index=False)
```
